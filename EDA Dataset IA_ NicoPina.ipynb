**Gráficas**

**H1: El uso intensivo de IAg baja la retención de conocimiento.**

```python

plt.figure(figsize=(10,6))
plt.scatter(df["Weekly_GenAI_Hours"],df["Skill_Retention_Score"],alpha=0.3,s=0.5, color="darkblue") # Configuramos el diagrama de dispersión entre "Weekly_GenAI_Hours" vs "Skill_Retention_Score".
sns.regplot(x=df["Weekly_GenAI_Hours"], y=df["Skill_Retention_Score"], scatter= False, line_kws={"color":"black", "linewidth":3})#Graficamos la linea de regresión para tener una idea global de como se relacional las variables.
plt.xlabel("Horas a la semana de uso de IAg", fontsize=12)
plt.ylabel("Retención del conocimiento", fontsize=12)
plt.title("Horas de uso semanal de IAg v/s Retención del conocimiento", fontsize=14)
plt.show()
```

*** H2: A menor uso semanal de la IAg, es mayor la cantidad de horas dedicadas al estudio tradicional (sin IAg)***

```python
plt.figure(figsize=(10,6))
plt.scatter(df["Weekly_GenAI_Hours"],df["Traditional_Study_Hours"],alpha=0.8,s=0.5, color="darkgreen") # Configuramos el diagrama de dispersión entre "Weekly_GenAI_Hours" vs"Traditional_Study_Hours" 
sns.regplot(x=df["Weekly_GenAI_Hours"], y=df["Traditional_Study_Hours"], scatter= False, line_kws={"color":"black", "linewidth":3}) #Graficamos la linea de regresión para tener una idea global de como se relacional las variables.
plt.xlabel("Horas a la semana de uso de IA")
plt.ylabel("Horas de estudio tradicional")
plt.title("Horas de uso Ia v/s Horas de estudio tradicional")

plt.show()
```

*** H3: Hay un aumento en el promedio de notas a mayor uso percibido de la IAg ***

```python
# Discretizamos con categorias la cantidad de horas de uso de la IAg semanal. 
df["Weekly_GenAI_Level"] = pd.cut(df["Weekly_GenAI_Hours"], bins= [0,5,10,20,100], labels=["Ocasional","Regular","Intensivo", "Extremo"], right=False )
```

```python
col = "Weekly_GenAI_Level"
cols = list(df.columns) # toma los nombres de las columnas y los convierte en una lista"
cols.remove(col) #Sacamos la columa creada "col" de la última posición.
cols.insert(5,col) # Insertamos la columna en en la posición 5. 
df=df[cols] # reordenamos el DataFrame.
```

```python
df.head() # Lo usamos nuevamente para verificar si se realizó el cambio.
```

```python
#Sacaremos el promedio Pre_Semester_GPA y Post_Semester_GPA  por cada categoria creada de Weekly_GenAI_Hours
prom_GPAs=df.groupby("Weekly_GenAI_Level")[["Pre_Semester_GPA", "Post_Semester_GPA"]].mean()
```

```python
# Redefinimos las etiquetas para incluir la cantidad de horas por categoría, de manera que pueda visualizarse de menor manera en el gráfico de barras dobles.
Etiquetas =["Ocasional\n[0,0 - 2,4 hrs]", "regular\n[2,4 - 5,8 hrs]", "Intensivo\n[5,8 - 11,7 hrs]",  "Extremo\n[11,7 - 40,0 hrs]"]
```

```python
plt.figure(figsize=(15,6))
ax=prom_GPAs.plot(kind="bar", color=["#07223d", "#55A868"])
plt.title("Comparación entre GPA promedio (Pre vs Post) con el Nivel de uso de la IA")
plt.xlabel("Nivel de uso de la IA")
plt.ylabel("GPA Promedio")
plt.legend(loc="center left", bbox_to_anchor=(1,0.5))   # cambiamos la posición de recuadro de leyendas. 
ax.set_xticklabels(Etiquetas, rotation=0)
plt.show()
```

P1: ¿La distribución del uso de IAg por semana es simétrica o está sesgada?

```python
plt.figure(figsize=(10,6))
sns.histplot(data=df["Weekly_GenAI_Hours"], color="#0EA5E9",edgecolor= "black", kde=True) #

#Calculamos los datos estadísticos relevantes Mediana y media de la variable "Weekly_GenAI_Hours".
media=df["Weekly_GenAI_Hours"].mean()
mediana=df["Weekly_GenAI_Hours"].median()
#------------------------------------------------------------------------------------------------
#Dibujamos lineas verticales marcando el valor de la media y la mediana. 
plt.axvline(media, color="black", linestyle="--", label=f"Media :{media} horas")
plt.axvline(mediana, color="red", linestyle=":", label=f"Mediana :{mediana} horas")
#------------------------------------------------------------------------------------------------
plt.title("Distribución de horas de semanales de uso de la IAg", fontsize=14)
plt.xlabel("Horas semanales de uso de la IAg", fontsize=12)
plt.ylabel("Número de Estudiantes", fontsize=12)
plt.legend()
plt.show()
```

**Cambio neto en GPA en relación al nivel de habilidad en promp engineering**

```python
df["Delta_GPA"] = df['Post_Semester_GPA'] - df['Pre_Semester_GPA'] # Calculamos el delta GPA como post_GPA - pre_GPA
plt.figure(figsize=(10,6))
Skill_order=["Beginner", "Intermediate", "Advanced"] 

sns.boxplot(data=df,
               x="Prompt_Engineering_Skill",
               y="Delta_GPA",
               hue="Prompt_Engineering_Skill",# Separa las variaciones según Skill_order y usa colores distintos según palette.
               order=Skill_order,
               palette=["#BB9283", "#E9D3B1", "#52314F"],
               linewidth=1.3
              )
plt.axhline(0,color="darkblue",linestyle="--", label="Sin cambio en GPA") # traza una linea demarcando el punto sin cambios de GPA ( deltraGPA=0)

plt.title("Densidad de crecimiento académico según el nivel de habilidad en ingeniería de Prompts", fontsize=14, fontweight='bold')
plt.xlabel("Nivel de habilidad en ingeniería de Prompts", fontsize=12)
plt.ylabel("Cambio neto en GPA (Post-Pre)", fontsize=12)
plt.tight_layout()
plt.show()            
```

**Matriz de correlación**

```python
corr=df.corr(numeric_only=True).round(2) # Calculamos el coeficiente de correlaciónde Pearson entre todas las columnas numéricas del DataFrame y lo redondeamos a dos decimales.and
```

```python
plt.figure(figsize=(8,7))
sns.heatmap(corr, annot=True, cmap="YlOrRd") # Grafica un mapa de calor poniendo el coeficiente de pearson ( por el annot=True) y un color por defecto.and
plt.title(" Matriz de correlación de columnas numéricas")
plt.show()
```

**Distribución de la retención del conocimiento de estudiantes que usan IAg mayor a 12 horas a la semana**

```python
df_alto_uso =df[df["Weekly_GenAI_Hours"] >= 12] # Aislamos la población crítica para aquellos que tiene hora de uso semana mayor a 12 horas.
```

```python
plt.figure(figsize=(15,6))
#contruimos el histograma con los datos de estudiantes que usan Iag más de 12 horas para estudiar la distribución del puntaje de retención de habilidades.

sns.histplot(data=df_alto_uso, x="Skill_Retention_Score", bins=50, kde=True, color="#07223D",edgecolor="#D1C03F", alpha=0.8) 
#Datos estadísticos relevantes (media y mediana) y  graficamos lineas verticales que marcan estas metricas.
media_Skill=df_alto_uso["Skill_Retention_Score"].mean()
mediana_Skill=df_alto_uso["Skill_Retention_Score"].median()

plt.axvline(media_Skill, color="black", linestyle="--", label=f"Media :{media} puntos") # Grafica la linea vertical que muestra la media.
plt.axvline(mediana_Skill, color="red", linestyle=":", label=f"Mediana :{mediana}puntos")  # Grafica la linea vertical que muestra la mediana.

plt.title("Distribución de Retención de conocimiento de usuarios con de IAg \n(> 12 hrs/semana)")
plt.xlabel("Puntaje de Retención de Habilidades", fontsize=13)
plt.xticks(range(0, 101, 5))
plt.ylabel("Número de Estudiantes",fontsize=13)
plt.legend()
plt.tight_layout()
plt.show()
```
